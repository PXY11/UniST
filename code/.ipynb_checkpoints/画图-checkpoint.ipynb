```python
from matplotlib import pyplot as plt
import pandas as pd
tb_timing = pd.read_csv('../data/tb_timing.csv',index_col=0)
tb_timing.plot(rot = 45,grid=True,legend=False,figsize=(8,3))
first_pos = tb_timing.index.tolist().index('2021-05-10 09:17:52')
second_pos = tb_timing.index.tolist().index('2021-07-25 22:57:11')
third_pos = tb_timing.index.tolist().index('2021-08-05 06:02:24')
plt.vlines(first_pos, 950000, 1230000, linestyles='-'
           , alpha=0.5, color='g', label='mean_over_ma')
plt.vlines(second_pos, 950000, 1230000, linestyles='dashed'
           , alpha=0.5, color='r', label='mean_over_ma')
plt.vlines(third_pos, 950000, 1230000, linestyles='dashed'
           , alpha=0.5, color='r', label='mean_over_ma')
plt.annotate('first pos begin', xy=(first_pos, 1230000), xytext=(first_pos+5000, 1230000),
             xycoords='data',arrowprops=dict(facecolor='white'))
plt.annotate('second pos begin', xy=(second_pos, 1230000), xytext=(second_pos-30000, 1230000),
             xycoords='data',arrowprops=dict(facecolor='white'))
plt.annotate('third pos begin', xy=(third_pos, 1230000), xytext=(third_pos+2000, 1130000),
             xycoords='data',arrowprops=dict(facecolor='white'))

plt.title('Total Balance')
plt.show()
```

```python
dataCoin_test_period_2 = pd.read_csv('../data/dataCoin_test_period_2.csv',index_col=0)
dataCoin_test_period_2.head()
dataCoin_test_period_2.set_index('datetime')['close'].plot(grid=True,figsize=(8,5),label='CLSOE')
dataCoin_test_period_2.set_index('datetime')['closeVma'].plot(grid=True,figsize=(8,5),rot=45,label='CLOSE_MA')

pos_1_begin = dataCoin_test_period_2.set_index('datetime').index.tolist().index('2021-05-10 09:00:00')
pos_1_end = dataCoin_test_period_2.set_index('datetime').index.tolist().index('2021-05-12 23:00:00')

pos_2_begin = dataCoin_test_period_2.set_index('datetime').index.tolist().index('2021-07-25 23:00:00')
pos_2_end = dataCoin_test_period_2.set_index('datetime').index.tolist().index('2021-08-05 06:00:00')

pos_3_begin = dataCoin_test_period_2.set_index('datetime').index.tolist().index('2021-08-05 07:00:00')
pos_3_end = len(dataCoin_test_period_2)

low = 2000
high = 4000
plt.vlines(pos_1_begin, low, high, linestyles='dashed', alpha=0.5, color='g', label='pos_1_begin')
plt.vlines(pos_1_end, low, high, linestyles='dashed', alpha=0.5, color='r', label='pos_1_end')
plt.vlines(pos_2_begin, low, high, linestyles='dashed', alpha=0.5, color='b', label='pos_2_begin')
plt.vlines(pos_2_end, low, high, linestyles='dashed', alpha=0.5, color='orange', label='pos_2_end')
plt.vlines(pos_3_begin, low, high, linestyles='dashed', alpha=0.5, color='pink', label='pos_3_begin')
plt.vlines(pos_3_end, low, high, linestyles='dashed', alpha=0.5, color='black', label='pos_3_end')

plt.annotate('pos_1_begin', xy=(pos_1_begin, 3800), xytext=(pos_1_begin, 4200),xycoords='data',arrowprops=dict(facecolor='white'))
plt.annotate('pos_1_end', xy=(pos_1_end, 3500), xytext=(pos_1_end+300, 3500),xycoords='data',arrowprops=dict(facecolor='white'))

plt.annotate('pos_2_begin', xy=(pos_2_begin, high), xytext=(pos_2_begin, 4300),xycoords='data',arrowprops=dict(facecolor='white'))
plt.annotate('pos_2_end', xy=(pos_2_end, high), xytext=(pos_2_end+300, 4000),xycoords='data',arrowprops=dict(facecolor='white'))

plt.annotate('pos_3_begin', xy=(pos_3_begin, 3500), xytext=(pos_3_begin+200, 3500),xycoords='data',arrowprops=dict(facecolor='white'))
plt.annotate('pos_3_end', xy=(pos_3_end, 2500), xytext=(pos_3_end-500, 2000),xycoords='data',arrowprops=dict(facecolor='white'))
plt.title('Close Price with MA50')
plt.show()
```
